#1.What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.
Anomaly Detection is the process of identifying data points, events, or patterns that deviate significantly from the norm in a dataset. These unusual instances are called anomalies or outliers, and detecting them is crucial in various domains like fraud detection, cybersecurity, health monitoring, and quality control.
Types of Anomalies
1.Point Anomaly
A point anomaly occurs when a single data point is significantly different from the rest of the data.
2.Contextual Anomaly (also called Conditional Anomaly)
A contextual anomaly is a data point that is considered anomalous in a specific context but may be normal in another.
Example:
A temperature of 30°C:
In summer, it might be normal.
In winter, it would be considered anomalous.
his is a contextual anomaly.
3.Collective Anomaly
A collective anomaly occurs when a group of related data points is anomalous with respect to the entire dataset, even if individual data points may not be.
Example:
In network traffic logs:
A sudden burst of high traffic during a normally low-activity time (e.g., midnight to 1 AM) could indicate a DDoS attack.
Individually, each traffic entry may look normal, but together, they form a collective anomaly.




#2.Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.
Comparison of Isolation Forest, DBSCAN, and Local Outlier Factor(LOF) in terms of their approach, assumptions, and suitable use cases:
1.Isolation Forest
Approach:
Based on the idea that anomalies are more susceptible to isolation.
Constructs an ensemble of random trees and isolates observations by randomly selecting features and split values.
The number of splits required to isolate a point determines its anomaly score (fewer splits = more anomalous).
Key Characteristics:
Unsupervised.
Efficient on high-dimensional datasets.
Linear time complexity with respect to data size.
Best Use Cases:
Large-scale anomaly detection in high-dimensional data.
Real-time detection tasks (e.g., fraud detection, intrusion detection).
When you want a model that is fast and scalable.
2.DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
Approach:
Groups points that are closely packed together (points with many nearby neighbors).
Marks points in low-density regions as outliers (noise).
Requires two parameters: epsilon (neighborhood radius) and minPts (minimum number of points in a neighborhood to form a cluster).
Key Characteristics:
Density-based clustering algorithm with outlier detection as a byproduct.
Does not assume any specific data distribution.
Struggles with varying densities and high-dimensional data.
Best Use Cases:
Detecting outliers in low- to medium-dimensional spatial data.
Applications like geospatial anomaly detection, image analysis, or sensor data.
Situations where clusters are irregularly shaped and clearly separated.
3.Local Outlier Factor (LOF)
Approach:
Measures the local deviation of a data point with respect to its neighbors.
An outlier is defined as having a lower density than its neighbors.
LOF score > 1 indicates potential outlier.
Key Characteristics:
Focuses on local density rather than global structure.
Good for identifying contextual anomalies.
Sensitive to parameter settings like number of neighbors (k).
Best Use Cases:
Contextual anomaly detection, especially in clusters of varying density.
Works well when you suspect some outliers are only outliers in their local context.
Suitable for low- to medium-dimensional datasets.




#3.What are the key components of a Time Series? Explain each with one example.
A Time Series is a sequence of data points collected or recorded at regular time intervals. The key components of a time series help in understanding the underlying patterns in the data. These components are:
1.Trend
Definition:
The long-term movement or direction in the data over time. It may be increasing, decreasing, or flat.
Example:
The average global temperature rising steadily over decades due to climate change.
2.Seasonality
Definition:
A repeating pattern or cycle in the data at regular intervals, often driven by calendar-related effects (e.g., months, quarters).
Example:
Retail sales increase every December due to holiday shopping.
3.Cyclicity(Cycle)
Definition:
Fluctuations in the data that are not of fixed period (unlike seasonality) and often tied to economic or natural cycles.
Example:
Economic recessions affecting stock market indices every few years—but not at regular intervals.
4.Irregular/Noise (Residual)
Definition:
The random variation or unpredictable fluctuations in the data that can't be explained by trend, seasonality, or cycle.
Example:
A sudden drop in website traffic due to a one-day server outage.

#4.Define Stationary in time series.How can you test and transform a non-stationary series into a stationary one?
A stationary time series is one whose statistical properties remain constant over time. This means:
Constant mean
Constant variance
Constant autocovariance (depends only on the lag, not time)
In other words, the time series does not show trends, seasonality, or changing volatility over time.
Most time series forecasting models (like ARIMA) assume stationarity because it's easier to model and make predictions on data that behaves consistently over time.
How to Test for Stationarity:
1.Visual Inspection:
Plot the time series.
Look for trends, seasonality, or changing variance.
If you see consistent upward/downward movement or periodic patterns, it's likely non-stationary.
2.Summary Statistics:
Divide the series into windows (e.g., early vs. late periods).
Compare mean and variance across windows.
Significant changes indicate non-stationarity.
3.Statistical Tests:
Augmented Dickey-Fuller (ADF) Test
Null hypothesis (H₀): The series is non-stationary (has a unit root).
If p-value < 0.05, reject H₀ → series is stationary.
KPSS Test (Kwiatkowski–Phillips–Schmidt–Shin)
Null hypothesis (H₀): The series is stationary.
If p-value < 0.05, reject H₀ → series is non-stationary.
How to Make a Series Stationary:
1.Differencing
-Subtract current value from the previous one:
Y′=Yt−Yt−1
​-Removes trend.
2.Seasonal Differencing
Subtract value from the same season in a previous cycle:
Y′=Yt−Yt−8
Removes seasonality (e.g.,𝑠=12 for monthly data).
3.Log or Power Transformation
Stabilizes variance.
Example:𝑌′=log(𝑌𝑡)
4.Detrending
Remove trend using regression or smoothing methods (e.g., polynomial fit, moving average)


#5.Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.
AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of their structure and applications in time series forecasting:
1.AR (AutoRegressive) Model
Structure:
Uses the past values of the time series to predict future values.
Expressed as:Yt=c+ϕ1Yt−1+ϕ2Yt−2+⋯+ϕpYt−p+ϵt
​Where:p = number of lag terms;
ϵt= white noise
Use Case:
Good when there’s strong autocorrelation in the data.
E.g:stock prices where today’s price is influenced by recent days.
2.MA (Moving Average) Model
Structure:
Uses past forecast errors to predict future values.
Expressed as:Yt=μ+ϵt+θ1ϵt−1+θ2ϵt−2+⋯+θqϵt−q
Where:q = number of lagged forecast errors
Use Case:
Effective when noise or random shocks impact future values.
3.ARIMA (AutoRegressive Integrated Moving Average)
Structure:
Combines AR and MA models with differencing to remove non-stationarity.
Expressed as ARIMA(p, d, q):
p = AR terms
d = number of differencing
q = MA terms
Use Case:
Used for non-seasonal time series with trend.
E.g., predicting sales data with a long-term trend.
4.SARIMA (Seasonal ARIMA)
Structure:
Extends ARIMA by adding seasonal components.
Expressed as ARIMA(p, d, q)(P, D, Q, s):
(P, D, Q) = seasonal components
s = seasonal period (e.g., 12 for monthly data with annual cycle)
Use Case:
For seasonal time series (e.g., monthly energy consumption, sales with yearly seasonality).
5.SARIMAX (Seasonal ARIMA with Exogenous Variables)
Structure:
SARIMA + exogenous variables (X).
Incorporates external regressors that might influence the time series.
Use Case:
Used when external variables (weather, promotions, economic indicators) impact the target series.
E.g., forecasting ice cream sales using both historical data and daily temperature.

```python
#8.Train a SARIMA model on the monthly airline passengers dataset.Forecast the next 12 months and visualize the results.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.seasonal import seasonal_decompose
# Load the dataset
df = pd.read_csv("AirPassengers.csv")
df.columns = ['Month', 'Passengers']
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)
# Optional: Decompose the series
decomposition = seasonal_decompose(df['Passengers'], model='multiplicative', period=12)
decomposition.plot()
plt.tight_layout()
plt.show()
# Train the SARIMA model
model = SARIMAX(df['Passengers'],
                order=(1, 1, 1),
                seasonal_order=(1, 1, 1, 12),
                enforce_stationarity=False,
                enforce_invertibility=False)

sarima_result = model.fit(disp=False)
# Forecast the next 12 months
forecast_steps = 12
forecast = sarima_result.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=df.index[-1] + pd.DateOffset(months=1), periods=forecast_steps, freq='MS')
forecast_mean = forecast.predicted_mean
forecast_conf_int = forecast.conf_int()
# Convert confidence intervals to numeric values
forecast_mean.index = forecast_index
forecast_conf_int.index = forecast_index
lower_series = forecast_conf_int.iloc[:, 0].astype(float).values
upper_series = forecast_conf_int.iloc[:, 1].astype(float).values
forecast_mean_values = forecast_mean.values
# Plot forecast
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Passengers'], label='Observed')
plt.plot(forecast_mean.index, forecast_mean_values, label='Forecast', color='darkgreen')
plt.fill_between(forecast_mean.index, lower_series, upper_series, color='lightgreen', alpha=0.4)
plt.title('SARIMA Forecast of Air Passengers (Next 12 Months)')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#9.Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
from sklearn.preprocessing import StandardScaler
# Load and prepare dataset
df = pd.read_csv("AirPassengers.csv")
df.columns = ['Month', 'Passengers']
df['Month'] = pd.to_datetime(df['Month'])
df.reset_index(drop=True, inplace=True)
df['MonthIndex'] = range(len(df))  # numeric time index
# Feature set: time index vs passengers
X = df[['MonthIndex', 'Passengers']]
# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Apply LOF
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
y_pred = lof.fit_predict(X_scaled)
# Detect outliers
df['Anomaly'] = y_pred
outliers = df[df['Anomaly'] == -1]
# Plot
plt.figure(figsize=(10, 6))
plt.scatter(df['MonthIndex'], df['Passengers'], color='blue', label='Normal')
plt.scatter(outliers['MonthIndex'], outliers['Passengers'], color='red', label='Outliers')
plt.title('Local Outlier Factor (LOF) Anomaly Detection')
plt.xlabel('Time Index')
plt.ylabel('Number of Passengers')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
